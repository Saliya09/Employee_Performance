### Domain Analysis

### Employee Information:
1. **EmpNumber:** A unique identifier for each employee.
2. **Age:** The age of the employee, indicating the demographic profile.
3. **Gender:** The gender of the employee.

### Education and Professional Background:
4. **EducationBackground:** The field of study or major in which the employee has an educational background.
5. **EmpEducationLevel:** The educational level achieved by the employee.

### Marital and Family Status:
6. **MaritalStatus:** Indicates whether the employee is single, married, or divorced.

### Job-related Information:
7. **EmpDepartment:** The department in the organization where the employee works.
8. **EmpJobRole:** The specific role or position the employee holds.
9. **BusinessTravelFrequency:** How often the employee travels for business purposes.
10. **DistanceFromHome:** The distance in miles from the employee's home to the workplace.
11. **EmpHourlyRate:** The hourly rate of pay for the employee.
12. **EmpJobInvolvement:** The level of involvement the employee has in their job.
13. **EmpJobLevel:** The hierarchical level of the employee's job.
14. **EmpJobSatisfaction:** The satisfaction level of the employee with their job.
15. **NumCompaniesWorked:** The number of companies the employee has worked for in the past.
16. **OverTime:** Whether the employee works overtime or not.
17. **EmpLastSalaryHikePercent:** The percentage increase in the employee's salary during the last hike.

### Work Environment and Relationships:
18. **EmpEnvironmentSatisfaction:** The satisfaction level of the employee with the work environment.
19. **EmpRelationshipSatisfaction:** The satisfaction level of the employee with work relationships.
20. **EmpWorkLifeBalance:** The satisfaction level of the employee with their work-life balance.

### Work Experience Information:
21. **TotalWorkExperienceInYears:** The total work experience of the employee.
22. **TrainingTimesLastYear:** The number of training sessions attended by the employee in the last year.
23. **ExperienceYearsAtThisCompany:** The number of years the employee has worked at the current company.
24. **ExperienceYearsInCurrentRole:** The number of years the employee has been in the current role.
25. **YearsSinceLastPromotion:** The number of years since the last promotion.
26. **YearsWithCurrManager:** The number of years the employee has been with the current manager.

### Outcomes and Performance:
27. **Attrition:** A binary indicator of whether the employee has left the company or not.
28. **PerformanceRating:** The performance rating assigned to the employee.

### Insights and Potential Analyses:
- **Attrition Prediction:** Explore factors contributing to employee attrition and build predictive models.
- **Performance Analysis:** Identify factors influencing employee performance ratings.
- **Job Satisfaction Study:** Analyze the relationship between job satisfaction and various employee attributes.
- **Work-Life Balance Impact:** Investigate how work-life balance satisfaction relates to other factors.
- **Career Progression Analysis:** Examine the impact of various factors on career progression within the company.
